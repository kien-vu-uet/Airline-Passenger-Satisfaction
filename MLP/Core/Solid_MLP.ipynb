# **I, Neural Network Architecture**

$Z_1 = [w_1 w_2 \dots w_{20}][x_1 x_2 \dots x_{20}]^T + b_1^1$

Một số hàm phi tuyến điển hình: $ReLu, Sigmoid, Tanh$.

$ReLu(X) = max(0, X)$

$A_1 = ReLu(Z_1)$

---


**1. Đầu ra với hàm kích hoạt Sigmoid**

Giả sử $P(y_i = 1 | x_i; W) = Sig(W^{(5)^T}A^{(4)} + B^{(4)})$

$\rightarrow P(y_i = 0| x_i; W) = 1 - Sig(W^{(5)^T}A^{(4)} + B^{(4)})$

$\rightarrow P(y_i | x_i; W) = Sig(W^{(5)^T}A^{(4)} + B^{(4)})^{y_i}(1-Sig(W^{(5)^T}A^{(4)} + B^{(4)}))^{1-y_i}$

Đồng thời xảy ra khớp giữa $N$ điểm dữ liệu $\implies$ Ta sẽ có 1 chuỗi xác xuất đồng thời với giả thiết từng điểm dữ liệu là độc lập:

$\begin{align}P(y | X; W) = \prod_{i = 1}^{N}p(y_i | x_i; W) = \prod_{i = 1}^{N}Sig(W^{(5)^T}A^{(4)} + B^{(4)})^{y_i}(1-Sig(W^{(5)^T}A^{(4)} + B^{(4)}))^{1-y_i}\end{align}$

Suy ra đi tìm $W$ để $W_{MLE} = argmax_W {P(y | X; W)}$

Lấy Logarit tự nhiên của hàm xác xuất sau đó đổi dấu, ta được một chuỗi tổng có $N$ thành phần, nên ta sẽ nhân tiếp với $\frac{1}{N}$

Ta được hàm mất mát Entropy chéo nhị phân.

*Với tham số tối ưu là $W$*

Ta cần tìm $W$ để hàm $J(W)$ sau có giá trị cực tiểu

$\begin{align}J(W) =  - \frac{1}{N} log p(y |X; W) = - \frac{1}{N} \sum_{i=1}^{N}(y_ilog(Sig(W^{(5)^T}A^{(4)} + B^{(4)})) + (1 - y_i)log(1-Sig(W^{(5)^T}A^{(4)} + B^{(4)})))\end{align}$

$\begin{align}\nabla_W J(W; X, y) = - \frac{1}{N} \sum_{i=1}^{N}(\frac{y_i}{Sig(W^{(5)^T}A^{(4)} + B^{(4)})} - \frac{1- y_i}{1 - Sig(W^{(5)^T}A^{(4)} + B^{(4)})}) \nabla_W Sig(W^{(5)^T}A^{(4)} + B^{(4)})\end{align}$

$\begin{align}\nabla_W J(W; X, y) = \frac{1}{N} \sum_{i=1}^{N}\frac{Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i}{Sig(W^{(5)^T}A^{(4)} + B^{(4)})(1 -Sig(W^{(5)^T}A^{(4)} + B^{(4)}))} \nabla_W Sig(W^{(5)^T}A^{(4)} + B^{(4)}) \\
 = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \nabla_W Z^{(5)} \end{align}$

*Với Tham số tối ưu là $B$*

$\begin{align}J(B) =  - \frac{1}{N} log p(y |X; B) = - \frac{1}{N} \sum_{i=1}^{N}(y_ilog(Sig(W^{(5)^T}A^{(4)} + B^{(4)})) + (1 - y_i)log(1-Sig(W^{(5)^T}A^{(4)} + B^{(4)})))\end{align}$

$\begin{align}\nabla_B J(B; X, y) = - \frac{1}{N} \sum_{i=1}^{N}(\frac{y_i}{Sig(W^{(5)^T}A^{(4)} + B^{(4)})} - \frac{1- y_i}{1 - Sig(W^{(5)^T}A^{(4)} + B^{(4)})}) \nabla_B Sig(W^{(5)^T}A^{(4)} + B^{(4)})\end{align}$

$\begin{align}\nabla_B J(B; X, y) = \frac{1}{N} \sum_{i=1}^{N}\frac{Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i}{Sig(W^{(5)^T}A^{(4)} + B^{(4)})(1 -Sig(W^{(5)^T}A^{(4)} + B^{(4)}))} \nabla_B Sig(W^{(5)^T}A^{(4)} + B^{(4)}) \\
 = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \nabla_B Z^{(5)} \end{align}$

---

**2. Tính toán Feed Forward (Lan truyền thuận)**

$A^{(0)} = X$

$Z^{(1)} = W^{(1)^T}X + B^{(0)}$

$A^{(1)} = ReLu(W^{(1)^T}X + B^{(0)})$

$Z^{(2)} = W^{(2)^T}A^{(1)} + B^{(1)}$

$A^{(2)} = ReLu(W^{(2)^T}A^{(1)} + B^{(1)})$

$Z^{(3)} = W^{(3)^T}A^{(2)} + B^{(2)}$

$A^{(3)} = ReLu(W^{(3)^T}A^{(2)} + B^{(2)})$

$Z^{(4)} = W^{(4)^T}A^{(3)} + B^{(3)}$

$A^{(4)} = ReLu(W^{(4)^T}A^{(3)} + B^{(3)})$

$Z^{(5)} = W^{(5)^T}A^{(4)} + B^{(4)}$

$A^{(5)} = Sig(W^{(5)^T}A^{(4)} + B^{(4)})$

---

**3. Tính toán Back Propagation (Lan truyền ngược)**

Tính gradient theo $W^{(5)}$ : 

$\begin{align}\nabla_{W^{(5)}} J(W; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \nabla_{W^{(5)}} Z^{(5)} =  \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) A^{(4)}\end{align}$

Tính gradient theo $B^{(4)}$ : 

$\begin{align}\nabla_{B^{(4)}} J(B; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \nabla_{B^{(4)}} Z^{(5)} =  \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \end{align}$

Tính gradient theo $W^{(4)}$ : 

$\begin{align}\nabla_{W^{(4)}} J(W; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \nabla_{W^{(4)}} Z^{(5)} \\
=  \dots  \nabla_{W^{(4)}} A^{(4)} \nabla_{A^{(4)}} Z^{(5)} = \dots  \nabla_{W^{(4)}} Z^{(4)} \nabla_{Z^{(4)}} A^{(4)}\nabla_{A^{(4)}} Z^{(5)}  \\
\Leftrightarrow \nabla_{W^{(4)}} J(w; x_i, y_i) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) A^{(3)}\big(\nabla_{Z(4)} ReLu(Z^{(4)})\big)W^{(5)^T}
\end{align}$

Tính gradient theo $B^{(3)}$ : 

$\begin{align}\nabla_{B^{(3)}} J(B; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \nabla_{B^{(3)}} Z^{(5)} \\
=  \dots  \nabla_{B^{(3)}} A^{(4)} \nabla_{A^{(4)}} Z^{(5)} = \dots  \nabla_{B^{(3)}} Z^{(4)} \nabla_{Z^{(4)}} A^{(4)}\nabla_{A^{(4)}} Z^{(5)}  \\
\Leftrightarrow \nabla_{B^{(3)}} J(w; x_i, y_i) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \big(\nabla_{Z(4)} ReLu(Z^{(4)})\big)W^{(5)^T}
\end{align}$

Tính gradient theo $W^{(3)}$ : 

$\begin{align}\nabla_{W^{(3)}} J(W; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i) \nabla_{W^{(3)}} Z^{(5)} \\
=  \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \nabla_{W^{(3)}} Z^{(3)} \nabla_{Z^{(3)}} A^{(3)} \nabla_{A^{(3)}} Z^{(4)} \nabla_{Z^{(4)}} A^{(4)} \nabla_{A^{(4)}} Z^{(5)} \\
\Leftrightarrow \nabla_{W^{(3)}} J(W; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i)A^{(2)} \big( \nabla_{Z(3)} ReLu(Z^{(3)})\big) W^{(4)^T} \big( \nabla_{Z(4)} ReLu(Z^{(4)}) \big) W^{(5)^T}
\end{align}$

Tính gradient theo $B^{(2)}$ : 

$\begin{align}\nabla_{B^{(2)}} J(B; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i) \nabla_{B^{(2)}} Z^{(5)} \\
=  \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \nabla_{B^{(2)}} Z^{(3)} \nabla_{Z^{(3)}} A^{(3)} \nabla_{A^{(3)}} Z^{(4)} \nabla_{Z^{(4)}} A^{(4)} \nabla_{A^{(4)}} Z^{(5)} \\
\Leftrightarrow \nabla_{B^{(2)}} J(B; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i) \big( \nabla_{Z(3)} ReLu(Z^{(3)})\big) W^{(4)^T} \big( \nabla_{Z(4)} ReLu(Z^{(4)}) \big) W^{(5)^T}
\end{align}$

Tính gradient theo $W^{(2)}$ : 

$\begin{align}\nabla_{W^{(2)}} J(W; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i) \nabla_{W^{(2)}} Z^{(5)} \\
=  \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \nabla_{W^{(2)}} Z^{(2)} \nabla_{Z^{(2)}} A^{(2)} \nabla_{A^{(2)}} Z^{(3)} \nabla_{Z^{(3)}} A^{(3)} \nabla_{A^{(3)}} Z^{(4)} \nabla_{Z^{(4)}} A^{(4)} \nabla_{A^{(4)}} Z^{(5)} \\
\Leftrightarrow \nabla_{W^{(2)}} J(W; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i)A^{(1)} \big( \nabla_{Z(2)} ReLu(Z^{(2)})\big) W^{(3)^T} \big( \nabla_{Z(3)} ReLu(Z^{(3)})\big) W^{(4)^T} \big( \nabla_{Z(4)} ReLu(Z^{(4)}) \big) W^{(5)^T}
\end{align}$

Tính gradient theo $B^{(1)}$ : 

$\begin{align}\nabla_{B^{(1)}} J(B; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i) \nabla_{B^{(1)}} Z^{(5)} \\
=  \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \nabla_{B^{(1)}} Z^{(2)} \nabla_{Z^{(2)}} A^{(2)} \nabla_{A^{(2)}} Z^{(3)} \nabla_{Z^{(3)}} A^{(3)} \nabla_{A^{(3)}} Z^{(4)} \nabla_{Z^{(4)}} A^{(4)} \nabla_{A^{(4)}} Z^{(5)} \\
\Leftrightarrow \nabla_{B^{(1)}} J(B; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i) \big( \nabla_{Z(2)} ReLu(Z^{(2)})\big) W^{(3)^T} \big( \nabla_{Z(3)} ReLu(Z^{(3)})\big) W^{(4)^T} \big( \nabla_{Z(4)} ReLu(Z^{(4)}) \big) W^{(5)^T}
\end{align}$

Tính gradient theo $W^{(1)}$ : 

$\begin{align}\nabla_{W^{(1)}} J(W; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i) \nabla_{W^{(1)}} Z^{(5)} \\
=  \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \nabla_{W^{(1)}} Z^{(1)} \nabla_{Z^{(1)}} A^{(1)} \nabla_{A^{(1)}} Z^{(2)} \nabla_{Z^{(2)}} A^{(2)} \nabla_{A^{(2)}} Z^{(3)} \nabla_{Z^{(3)}} A^{(3)} \nabla_{A^{(3)}} Z^{(4)} \nabla_{Z^{(4)}} A^{(4)} \nabla_{A^{(4)}} Z^{(5)} \\
\Leftrightarrow \nabla_{W^{(1)}} J(W; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i) X \big( \nabla_{Z(1)} ReLu(Z^{(1)})\big) W^{(2)^T} \big( \nabla_{Z(2)} ReLu(Z^{(2)})\big) W^{(3)^T} \big( \nabla_{Z(3)} ReLu(Z^{(3)})\big) W^{(4)^T} \big( \nabla_{Z(4)} ReLu(Z^{(4)}) \big) W^{(5)^T}
\end{align}$

Tính gradient theo $B^{(0)}$ : 

$\begin{align}\nabla_{B^{(0)}} J(B; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i) \nabla_{B^{(0)}} Z^{(5)} \\
=  \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i
) \nabla_{B^{(0)}} Z^{(1)} \nabla_{Z^{(1)}} A^{(1)} \nabla_{A^{(1)}} Z^{(2)} \nabla_{Z^{(2)}} A^{(2)} \nabla_{A^{(2)}} Z^{(3)} \nabla_{Z^{(3)}} A^{(3)} \nabla_{A^{(3)}} Z^{(4)} \nabla_{Z^{(4)}} A^{(4)} \nabla_{A^{(4)}} Z^{(5)} \\
\Leftrightarrow \nabla_{B^{(0)}} J(B; X, y) = \frac{1}{N} \sum_{i=1}^{N}(Sig(W^{(5)^T}A^{(4)} + B^{(4)}) - y_i) \big( \nabla_{Z(1)} ReLu(Z^{(1)})\big) W^{(2)^T} \big( \nabla_{Z(2)} ReLu(Z^{(2)})\big) W^{(3)^T} \big( \nabla_{Z(3)} ReLu(Z^{(3)})\big) W^{(4)^T} \big( \nabla_{Z(4)} ReLu(Z^{(4)}) \big) W^{(5)^T}
\end{align}$

\\
Nhận thấy nếu tính toán gradient từ tầng thứ 5 đầu tiên thì ta sẽ sử dụng được lại chuỗi gradient để tính cho các tầng trước đó. Vậy nên ta sẽ tính ngược đạo hàm từ tầng cuối cùng lên tầng đầu vào, Đó là thuật toán Back Propagation.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42) # set default random
```

```python
def Binary_Cross_Entropy_Loss(A2, Y):    
    m = Y.shape[0] # number of Samples
    # Compute the binary cross-entropy loss
    logprobs = np.dot(Y.T,np.log(A2 + 1e-8)) + np.dot((1-Y).T,np.log((1-A2+1e-8))) # norm penalties
    cost = -logprobs/m
    cost = float(np.squeeze(cost))  # makes sure cost is the dimension we expect.
    return cost
```

```python
class ourMLP:
    def __init__(self):
        self.W1 = np.random.randn(20,512) * 0.01
        self.B0 = np.zeros(512)
        # Drop Out 0.3
        self.W2 = np.random.randn(512,256) * 0.01
        self.B1 = np.zeros(256)
        self.W3 = np.random.randn(256,256) * 0.01
        self.B2 = np.zeros(256)
        # Drop Out 0.3
        self.W4 = np.random.randn(256,128) * 0.01
        self.B3 = np.zeros(128)
        self.W5 = np.random.randn(128,1) * 0.01
        self.B4 = np.zeros(1)
        self.lost_hist = []

    def fit(self, X, y, batch_size=32, validation_data={'y_val': None, 'X_val': None}, restore_best_weights=True, patience=10, epochs=50, dropout=True):
        # init hyperparamters for adam optimizer
        #######################################
        m_W1, v_W1 = np.zeros((X.shape[1], 1)), np.zeros((X.shape[1], 1))
        alpha_W1, beta1_W1, beta2_W1, eps_W1= 0.001, 0.9, 0.999, 1e-8

        m_W2, v_W2 = np.zeros((512, 1)), np.zeros((512, 1))
        alpha_W2, beta1_W2, beta2_W2, eps_W2= 0.001, 0.9, 0.999, 1e-8

        m_W3, v_W3 = np.zeros((256, 1)), np.zeros((256, 1))
        alpha_W3, beta1_W3, beta2_W3, eps_W3= 0.001, 0.9, 0.999, 1e-8

        m_W4, v_W4 = np.zeros((256, 1)), np.zeros((256, 1))
        alpha_W4, beta1_W4, beta2_W4, eps_W4= 0.001, 0.9, 0.999, 1e-8

        m_W5, v_W5 = np.zeros((128, 1)), np.zeros((128, 1))
        alpha_W5, beta1_W5, beta2_W5, eps_W5= 0.001, 0.9, 0.999, 1e-8

        m_B0, v_B0 = np.zeros(512), np.zeros(512)
        alpha_B0, beta1_B0, beta2_B0, eps_B0= 0.001, 0.9, 0.999, 1e-8

        m_B1, v_B1 = np.zeros(256), np.zeros(256)
        alpha_B1, beta1_B1, beta2_B1, eps_B1= 0.001, 0.9, 0.999, 1e-8

        m_B2, v_B2 = np.zeros(256), np.zeros(256)
        alpha_B2, beta1_B2, beta2_B2, eps_B2= 0.001, 0.9, 0.999, 1e-8

        m_B3, v_B3 = np.zeros(128), np.zeros(128)
        alpha_B3, beta1_B3, beta2_B3, eps_B3= 0.001, 0.9, 0.999, 1e-8

        m_B4, v_B4 = np.zeros(1), np.zeros(1)
        alpha_B4, beta1_B4, beta2_B4, eps_B4= 0.001, 0.9, 0.999, 1e-8
        #######################################

        weight_hist = [] # to store best weight
        N = X.shape[0]
        nbatches = int(np.ceil(float(N)/batch_size))
        
        for ep in range(epochs):
            mix_idxs = np.random.permutation(N) # Stochastic shuffle dataset
            for i in range (nbatches):
                batch_idxs = mix_idxs[batch_size*i : min(batch_size*(i+1), N)]
                X_batch, y_batch = X[batch_idxs], y[batch_idxs].reshape(-1, 1)
                
                # Feed Forward
                Z1 = X_batch.dot(self.W1) + self.B0
                A1 = Z1.copy()
                A1[A1 < 0] = 0 # ReLu
                # Dropout keep probability = 0.7
                D1 = np.random.rand(A1.shape[0], A1.shape[1])
                D1 = D1 < 0.7
                A1 = A1 * D1
                A1 = A1 / 0.7
                # end drop
                Z2 = A1.dot(self.W2) + self.B1
                A2 = Z2.copy()
                A2[A2 < 0] = 0 # ReLu
                Z3 = A2.dot(self.W3) + self.B2
                A3 = Z3.copy()
                A3[A3 < 0] = 0 # ReLu
                # Dropout keep probability = 0.7
                D3 = np.random.rand(A3.shape[0], A3.shape[1])
                D3 = D3 < 0.7
                A3 = A3 * D3
                A3 = A3 / 0.7
                # end drop
                Z4 = A3.dot(self.W4) + self.B3
                A4 = Z4.copy()
                A4[A4 < 0] = 0 # ReLu
                Z5 = A4.dot(self.W5) + self.B4
                A5 = 1 / (1 + np.exp(-Z5)) # sigmoid activate function

                # print lost value after each 5 epochs
                if i == nbatches - 1:
                    # Cross Validation with validation data
                    loss = Binary_Cross_Entropy_Loss(self.predict(validation_data['X_val']), validation_data['y_val'].reshape(-1, 1))
                    if ep % 5 == 0:
                        print("epoch: %d, validation loss: %f" % (ep, loss))
                    self.lost_hist.append(loss)
                    weight_hist.append({'W1':self.W1, 'W2':self.W2, 'W3':self.W3, 'W4':self.W4, 'W5':self.W5 \
                                        , 'B0':self.B0, 'B1':self.B1, 'B2':self.B2, 'B3':self.B3, 'B4':self.B4})    
                    
                '''
                X_batch: N x 20
                W1: 20 x 512, X: N x 20, B0: 512, Z1: N x 512, A1: N x 512
                W2: 512 x 256, Z2: N x 256, A2: N x 256, B1: 256
                W3: 256 x 256, Z3: N x 256, A3: N x 256, B2: 256
                W4: 256 x 128, Z4: N x 128, A4: N x 128, B3: 128
                W5: 128 x 1, Z5: N x 1, A5: N x 1, B4: 1 x 1
                '''
                #######################################
                # Back Propagation
                # 5th layer
                dZ5 = A5 - y_batch
                dW5 = (1/X_batch.shape[0]) * A4.T.dot(dZ5)
                dB4 = (1/X_batch.shape[0]) * np.sum(dZ5, axis = 0)
                
                # 4th layer
                dZ4 = dZ5.dot(self.W5.T)
                dZ4[Z4 <= 0] = 0
                dB3 = (1/X_batch.shape[0]) * np.sum(dZ4, axis = 0)
                dW4 = (1/X_batch.shape[0]) * A3.T.dot(dZ4)

                # 3rd layer
                dZ3 = dZ4.dot(self.W4.T)
                # Drop out corresponding
                dZ3 = dZ3 * D3
                dZ3 = dZ3 / 0.7
                # end
                dZ3[Z3 <= 0] = 0
                dB2 = (1/X_batch.shape[0]) * np.sum(dZ3, axis = 0)
                dW3 = (1/X_batch.shape[0]) * A2.T.dot(dZ3)

                # 2nd layer
                dZ2 = dZ3.dot(self.W3.T)
                dZ2[Z2 <= 0] = 0
                dB1 = (1/X_batch.shape[0]) * np.sum(dZ2, axis = 0)
                dW2 = (1/X_batch.shape[0]) * A1.T.dot(dZ2)

                # 1st layer
                dZ1 = dZ2.dot(self.W2.T)
                # Drop out corresponding
                dZ1 = dZ1 * D1
                dZ1 = dZ1 / 0.7
                # end
                dZ1[Z1 <= 0] = 0
                dB0 = (1/X_batch.shape[0]) * np.sum(dZ1, axis = 0)
                dW1 = (1/X_batch.shape[0]) * X_batch.T.dot(dZ1)
                #######################################

                #######################################
                # Optimizer sADAM
                # Stochastic AdaGrad Momentum Gradient Descent
                # for W1
                # calculate momentum 1
                m_W1 = beta1_W1 * m_W1 + (1.0 - beta1_W1) * dW1
                # calculate momentum 2
                v_W1 = beta2_W1 * v_W1 + (1.0 - beta2_W1) * dW1**2
                mhat_W1 = m_W1 / (1.0 - beta1_W1**(ep + 1))
                vhat_W1 = v_W1 / (1.0 - beta2_W1**(ep + 1))
                self.W1 = self.W1 - alpha_W1 * mhat_W1 / (np.sqrt(vhat_W1) + eps_W1)
                # self.W1 = self.W1 - alpha * dW1 -- Simple Stochastic Gradient Descent


                # for W2
                # calculate momentum 1
                m_W2 = beta1_W2 * m_W2 + (1.0 - beta1_W2) * dW2
                # calculate momentum 2
                v_W2 = beta2_W2 * v_W2 + (1.0 - beta2_W2) * dW2**2
                mhat_W2 = m_W2 / (1.0 - beta1_W2**(ep + 1))
                vhat_W2 = v_W2 / (1.0 - beta2_W2**(ep + 1))
                self.W2 = self.W2 - alpha_W2 * mhat_W2 / (np.sqrt(vhat_W2) + eps_W2)
                # self.W2 = self.W2 - alpha * dW2 -- Simple Stochastic Gradient Descent

                # for W3
                # calculate momentum 1
                m_W3 = beta1_W3 * m_W3 + (1.0 - beta1_W3) * dW3
                # calculate momentum 2
                v_W3 = beta2_W3 * v_W3 + (1.0 - beta2_W3) * dW3**2
                mhat_W3 = m_W3 / (1.0 - beta1_W3**(ep + 1))
                vhat_W3 = v_W3 / (1.0 - beta2_W3**(ep + 1))
                self.W3 = self.W3 - alpha_W3 * mhat_W3 / (np.sqrt(vhat_W3) + eps_W3)
                # self.W3 = self.W3 - alpha * dW3 -- Simple Stochastic Gradient Descent

                # for W4
                # calculate momentum 1
                m_W4 = beta1_W4 * m_W4 + (1.0 - beta1_W4) * dW4
                # calculate momentum 2
                v_W4 = beta2_W4 * v_W4 + (1.0 - beta2_W4) * dW4**2
                mhat_W4 = m_W4 / (1.0 - beta1_W4**(ep + 1))
                vhat_W4 = v_W4 / (1.0 - beta2_W4**(ep + 1))
                self.W4 = self.W4 - alpha_W4 * mhat_W4 / (np.sqrt(vhat_W4) + eps_W4)
                # self.W4 = self.W4 - alpha * dW4 -- Simple Stochastic Gradient Descent

                # for W5
                # calculate momentum 1
                m_W5 = beta1_W5 * m_W5 + (1.0 - beta1_W5) * dW5
                # calculate momentum 2
                v_W5 = beta2_W5 * v_W5 + (1.0 - beta2_W5) * dW5**2
                mhat_W5 = m_W5 / (1.0 - beta1_W5**(ep + 1))
                vhat_W5 = v_W5 / (1.0 - beta2_W5**(ep + 1))
                self.W5 = self.W5 - alpha_W5 * mhat_W5 / (np.sqrt(vhat_W5) + eps_W5)
                # self.W5 = self.W5 - alpha * dW5 -- Simple Stochastic Gradient Descent

                # for B0
                # calculate momentum 1
                m_B0 = beta1_B0 * m_B0 + (1.0 - beta1_B0) * dB0
                # calculate momentum 2
                v_B0 = beta2_B0 * v_B0 + (1.0 - beta2_B0) * dB0**2
                mhat_B0 = m_B0 / (1.0 - beta1_B0**(ep + 1))
                vhat_B0 = v_B0 / (1.0 - beta2_B0**(ep + 1))
                self.B0 = self.B0 - alpha_B0 * mhat_B0 / (np.sqrt(vhat_B0) + eps_B0)
                # self.B0 = self.B0 - alpha * dB0 -- Simple Stochastic Gradient Descent

                # for B1
                # calculate momentum 1
                m_B1 = beta1_B1 * m_B1 + (1.0 - beta1_B1) * dB1
                # calculate momentum 2
                v_B1 = beta2_B1 * v_B1 + (1.0 - beta2_B1) * dB1**2
                mhat_B1 = m_B1 / (1.0 - beta1_B1**(ep + 1))
                vhat_B1 = v_B1 / (1.0 - beta2_B1**(ep + 1))
                self.B1 = self.B1 - alpha_B1 * mhat_B1 / (np.sqrt(vhat_B1) + eps_B1)
                # self.B1 = self.B1 - alpha * dB1 -- Simple Stochastic Gradient Descent

                # for B2
                # calculate momentum 1
                m_B2 = beta1_B2 * m_B2 + (1.0 - beta1_B2) * dB2
                # calculate momentum 2
                v_B2 = beta2_B2 * v_B2 + (1.0 - beta2_B2) * dB2**2
                mhat_B2 = m_B2 / (1.0 - beta1_B2**(ep + 1))
                vhat_B2 = v_B2 / (1.0 - beta2_B2**(ep + 1))
                self.B2 = self.B2 - alpha_B2 * mhat_B2 / (np.sqrt(vhat_B2) + eps_B2)
                # self.B2 = self.B2 - alpha * dB2 -- Simple Stochastic Gradient Descent

                # for B3
                # calculate momentum 1
                m_B3 = beta1_B3 * m_B3 + (1.0 - beta1_B3) * dB3
                # calculate momentum 2
                v_B3 = beta2_B3 * v_B3 + (1.0 - beta2_B3) * dB3**2
                mhat_B3 = m_B3 / (1.0 - beta1_B3**(ep + 1))
                vhat_B3 = v_B3 / (1.0 - beta2_B3**(ep + 1))
                self.B3 = self.B3 - alpha_B3 * mhat_B3 / (np.sqrt(vhat_B3) + eps_B3)
                # self.B3 = self.B3 - alpha * dB3 -- Simple Stochastic Gradient Descent

                # for B4
                # calculate momentum 1
                m_B4 = beta1_B4 * m_B4 + (1.0 - beta1_B4) * dB4
                # calculate momentum 2
                v_B4 = beta2_B4 * v_B4 + (1.0 - beta2_B4) * dB4**2
                mhat_B4 = m_B4 / (1.0 - beta1_B4**(ep + 1))
                vhat_B4 = v_B4 / (1.0 - beta2_B4**(ep + 1))
                self.B4 = self.B4 - alpha_B4 * mhat_B4 / (np.sqrt(vhat_B4) + eps_B4)
                # self.B4 = self.B4 - alpha * dB4 -- Simple Stochastic Gradient Descent
                self.B4 = self.B4 - alpha_B4 * dB4
                #######################################
        
        # store best weight
        opt_weight = weight_hist[self.lost_hist.index(min(self.lost_hist))]
        self.W1, self.W2, self.W3, self.W4, self.W5, self.B0, self.B1, self.B2, self.B3, self.B4 = \
        opt_weight['W1'], opt_weight['W2'], opt_weight['W3'], opt_weight['W4'], opt_weight['W5'],\
         opt_weight['B0'], opt_weight['B1'], opt_weight['B2'], opt_weight['B3'], opt_weight['B4']
        

    def predict(self, X):
        Z1 = X.dot(self.W1) + self.B0
        A1 = Z1
        A1[A1 < 0] = 0 # ReLu
        Z2 = A1.dot(self.W2) + self.B1
        A2 = Z2 
        A2[A2 < 0] = 0 # ReLu
        Z3 = A2.dot(self.W3) + self.B2
        A3 = Z3
        A3[A3 < 0] = 0 # ReLu
        Z4 = A3.dot(self.W4) + self.B3
        A4 = Z4
        A4[A4 < 0] = 0 # ReLu
        Z5 = A4.dot(self.W5) + self.B4
        return 1 / (1 + np.exp(-Z5))

    def save_weight(self):
        np.savetxt('/content/drive/My Drive/Colab Notebooks/AI Project/MLP/Parameters/W1.txt', self.W1)
        np.savetxt('/content/drive/My Drive/Colab Notebooks/AI Project/MLP/Parameters/W2.txt', self.W2)
        np.savetxt('/content/drive/My Drive/Colab Notebooks/AI Project/MLP/Parameters/W3.txt', self.W3)
        np.savetxt('/content/drive/My Drive/Colab Notebooks/AI Project/MLP/Parameters/W4.txt', self.W4)
        np.savetxt('/content/drive/My Drive/Colab Notebooks/AI Project/MLP/Parameters/W5.txt', self.W5)
        np.savetxt('/content/drive/My Drive/Colab Notebooks/AI Project/MLP/Parameters/B0.txt', self.B0)
        np.savetxt('/content/drive/My Drive/Colab Notebooks/AI Project/MLP/Parameters/B1.txt', self.B1)
        np.savetxt('/content/drive/My Drive/Colab Notebooks/AI Project/MLP/Parameters/B2.txt', self.B2)
        np.savetxt('/content/drive/My Drive/Colab Notebooks/AI Project/MLP/Parameters/B3.txt', self.B3)
        np.savetxt('/content/drive/My Drive/Colab Notebooks/AI Project/MLP/Parameters/B4.txt', self.B4)

    def load_pretrained(self, weight_collection):
        self.W1 = weight_collection['W1']
        self.W2 = weight_collection['W2']
        self.W3 = weight_collection['W3']
        self.W4 = weight_collection['W4']
        self.W5 = weight_collection['W5']
        self.B0 = weight_collection['B0']
        self.B1 = weight_collection['B1']
        self.B2 = weight_collection['B2']
        self.B3 = weight_collection['B3']
        self.B4 = weight_collection['B4']
```
