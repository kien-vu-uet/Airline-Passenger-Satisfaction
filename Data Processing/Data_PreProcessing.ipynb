# **1. ENCODING**

## **Phải xử lý các trường Object trước khi đưa vào mô hình**

Có các loại Encoding sau:

- Onehot encoding: Bị lời nguyền về số chiều dữ liệu

- Hash encoding: Phải đảm bảo có thể decode ngược lại

- Frequency encoding: Sẽ là vô nghĩa nếu dữ liệu có phân phối đều

- Label encoding: Sẽ vô tình thêm nhiễu vào dữ liệu, giả sử nếu dữ liệu không có tính ưu tiên thì việc đánh số dữ liệu sẽ sinh ra nhiễu ưu tiên.

Các trường cần được encoding bên trong dataframe: Gender, Customer Type, Type of Travel, Class, satisfaction (label)

```python
train_df.nunique()
```

Một vài trường thuộc tính có số lượng giá trị unique là 2 hoặc 3. Ta có thể dùng labels encoding để encode các trường đó

Thực hiện encode 5 thuộc tính dạng chữ của 2 bộ train và test

```python
[col for col in train_df.columns if train_df[col].dtype == 'object']
```

```python
print(train_df['Gender'].unique())
print(train_df['Customer Type'].unique())
print(train_df['Type of Travel'].unique())
print(train_df['Class'].unique())
print(train_df['satisfaction'].unique())
```

```python
# Encode columns
train_df['Gender'] = train_df['Gender'].map({'Male' : 0, 'Female' : 1})
test_df['Gender'] = test_df['Gender'].map({'Male' : 0, 'Female' : 1})

train_df['Customer Type'] = train_df['Customer Type'].map({'Loyal Customer' : 0, 'disloyal Customer' : 1})
test_df['Customer Type'] = test_df['Customer Type'].map({'Loyal Customer' : 0, 'disloyal Customer' : 1})

train_df['Type of Travel'] = train_df['Type of Travel'].map({'Personal Travel' : 0, 'Business travel' : 1})
test_df['Type of Travel'] = test_df['Type of Travel'].map({'Personal Travel' : 0, 'Business travel' : 1})

train_df['Class'] = train_df['Class'].map({'Eco' : 0, 'Eco Plus' : 1, 'Business' : 2})
test_df['Class'] = test_df['Class'].map({'Eco' : 0, 'Eco Plus' : 1, 'Business' : 2})

train_df['satisfaction'] = train_df['satisfaction'].map({'neutral or dissatisfied' : 0, 'satisfied' : 1})
test_df['satisfaction'] = test_df['satisfaction'].map({'neutral or dissatisfied' : 0, 'satisfied' : 1})
```

3.1. Chuẩn hóa các trường có thang đo cố định (5)

```python
def standardRating(df, cols):
  # S is a series
  for col in cols:
    m = np.max(df[col])
    df[col] = df[col].apply(lambda x: x / m)
    
  return df
```

```python
pc1 = ['Ease of Online booking', 'Inflight wifi service'] #k=1
pc2 = ['Departure Delay in Minutes', 'Arrival Delay in Minutes']#k=1
```

```python
from sklearn.decomposition import PCA
```

```python
def reduce_dim(df, col, idx):
  pca = PCA(n_components = 1)
  pca.fit(df[col])
  pca_red = pca.transform(df[col])
  pca_df = pd.DataFrame(data = pca_red, columns = ['pca_f' + str(idx)])
  df = pd.concat([pca_df, df], ignore_index = False, sort = False, axis = 1)
  df.drop(col, axis = 1, inplace=True)
  print(np.cumsum(pca.explained_variance_ratio_))
  return df
```

```python
train_df = reduce_dim(train_df.copy(), pc1, 0)
train_df = reduce_dim(train_df.copy(), pc2, 1)
train_df[['pca_f0', 'pca_f1']]
```
