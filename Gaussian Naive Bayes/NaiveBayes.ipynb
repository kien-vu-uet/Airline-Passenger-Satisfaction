# **Naive Bayes**

**1. Khái niệm về mô hình phân lớp**

*Mô hình phân lớp* là một mô hình Machine Learning dùng để phân loại các vật mẫu dựa trên các đặc tính đã xác định.

**2.Mô hình phân lớp Navie Bayes**

Naive Bayes là một thuật toán phân lớp được mô hình hóa dựa trên Bayes trong xác suất thống kê:

$$P(y \mid X) =\frac{P(y) P(X\mid y)}{P(X)} $$

Trong đó: 



*   $P(y \mid X)$ được gọi là Posterior Probabillity: xác suất của mục tiêu y với điều kiện có đặc trưng X
*   $P(X \mid y)$ dược gọi là likelihood: xác suất của đặc trưng X khi đã biết mục tiêu y

*   $P(y)$ được gọi là prior probabillity của mục tiêu y
*   $P(X)$ được gọi là prior probabillity của mục tiêu X

Ở đây, X là vecto các đặc trưng, có thể viết dưới dạng: 
$ X = (x_1, x_2, x_3,..., x_n)$
Khi đó, đẳng thức Bayes trở thành:
$$P (y \mid x_1,..., x_n)=\frac{P (y) P (x_1,..., x_n \mid y)}{P (x_1,..., x_n)} $$
Đẳng thức Bayes tương đương với:
$$P (y \mid x_1,..., x_n)=\frac{P (y) P(x_1\mid y)P(x_2\mid y)...P(x_n\mid y)}{P (x_1)P (x_2)...P (x_n)} $$


Trong mô hình Naive Bayes, có 2 giá thiết được đặt ra:


1.   Các đặc trưng đưa vào mô hình là độc lập với nhau. Tức là sự thay đổi giá trị của một đặc trưng không ảnh hưởng đến các đặc trưng còn lại
2.   Các đặc trưng đưa vào mô hình có ảnh hưởng ngang nhau đối với đầu ra mục tiêu

Khi đó, kết quả mục tiêu y để $P(y \mid X)$ đạt cực đại trở thành 
$$̂{y} =\ arg \ max_y P (y) \prod_{i=1}^{n} P (x_i \mid y) $$

Chính vì hai giả thiết gần như không tồn tại trong thực tế trên, mô hình này mới được gọi là Naive (ngây thơ). Tuy nhiên, chính sự đơn giản của nó với việc dự đoán rất nhanh kết quả đầu ra khiến nó được sử dụng rất nhiều trong thực tế trên những bộ dữ liệu lớn, đem lại kết quả khả quan. Một vài ứng dụng của Naive Bayes có thể kể đến như: lọc thư rác, phân loại văn bản, dư đoán sắc thái văn bản, ...

**3. Một số kiểu mô hình Naive Bayes**

3.1 Multinomial Naive Bayes

Mô hình này chủ yếu được sử dụng trong phân loại văn bản. Đặc trưng đầu vào ở đây chính là dữ liệu thường được biểu thị dưới dạng số lượng vecto từ, mặc dù vecto tf-idf cũng được biết là hoạt động tốt trong thực tế. Phân phối được tham số hóa bởi các vecto $\theta_y = (\theta_ {y1}, \ldots, \theta_ {yn}) $ cho mỗi lớp y, trong đó n là số lượng các tính năng (trong phân loại văn bản, kích thước của từ vựng) và $\theta_ {yi}$ là xác suất của $P(x_i\mid y)$ của tính năng i xuất hiện trong một mẫu thuộc về lớp y

Những thông số $\theta_ {y}$ được ước tính bằng công thức đếm tần số tương đối:
$$\hat {\theta} _ {yi} =\frac {N_ {yi} + \alpha} {N_y + \alpha n}$$
Trong đó: 

$N_{yi} = \sum_{x \in T} x_i$ là số lần tính năng của i xuất hiện trong một lớp mẫu y trong tập T

$N_{y} = \sum_{i=1}^{n} N_{yi}$ là tổng số tất cả các tính năng trong lớp y

3.2 Bernoulli Naive Bayes 

Mô hình này được sử dụng cho dữ liệu rời rạc và nó hoạt động trên phân phối Bernoulli. Tính năng chính của Bernoulli Navie Bayes là nó chỉ chấp nhận các đặc điểm dưới dạng giá trị nhị phân như đúng hoặc sai, có hoặc không, thành công hay thất bại, 0 hoặc 1, v.v. Vì vậy, khi các giá trị đặc trưng là nhị phân, chúng ta biết răng chúng ta phải sử dụng bộ phân loại Bernoulli Naive Bayes.

Quy tắc quyết định cho Bernoulli ngây thơ Bayes dựa trên

$$P (x_i \mid y) = P (i \mid y) x_i +(1 - P (i \mid y)) (1 - x_i) $$
khác với quy tắc Multinomial Naive Bayes ở chỗ nó xử phạt rõ ràng sự không xuất hiện của một tính năng đó là một chỉ số cho lớp học, trong đó biến thể đa thức sẽ đơn giản bỏ qua một đặc điểm không xảy ra.

3.3 Gausian Naive Bayes

Các hàm khác có thể sử dụng để ước tính phân phối dữ liệu, nhưng Gaussian (hoặc phân bố chuẩn) là dễ nhất để làm việc vì bạn chỉ cần ước tính giá trị trung bình và độ lệch chuẩn từ dữ liệu đào tạo của bạn. Điều này có nghĩa là ngoài các xác suất của mỗi lớp, chúng ta cũng phải lưu trữ độ lệch trung bình và độ lệch chuẩn cho từng biến đầu vào mỗi lớp.

Khả năng các tính năng được giả định là Gausian:
$$P(x_i \mid y) = \frac {1} {\sqrt {2 \pi \sigma ^ 2_y}} \exp (- \frac {(x_i - \mu_y) ^ 2} {2 \sigma ^ 2_y })$$

Trong đó 

*   $\sigma_y$ là độ lệch chuẩn trong XSTK
*   $\mu_y$ là là giá trị kì vọng của biến ngẫu nhiên

1. Viết các hàm trọn gói trong 1 class

```python
class GaussianNaiveBayes:
    def __init__(self, n_samples = 1000, n_features = 10, n_classes = 2):
        self.n_samples = n_samples
        self.n_features = n_features
        self.n_classes = n_classes

    def fit(self, X, y):
        # get number of samples (rows) and features (columns)
        self.n_samples, self.n_features = X.shape
        # get number of uniques classes
        self.n_classes = len(np.unique(y))

        # create three zero-matrices to store summary stats & prior
        self.mean = np.zeros((self.n_classes, self.n_features))
        self.variance = np.zeros((self.n_classes, self.n_features))
        self.priors = np.zeros(self.n_classes)

        # remain the samples
        self.X = X
        self.y = y

        for c in range(self.n_classes):
            # create a subset of data for the specific class 'c'
            X_c = X[y == c]
            
            # calculate statistics and update zero-matrices, rows=classes, cols=features
            self.mean[c, :] = np.mean(X_c, axis=0)
            self.variance[c, :] = np.var(X_c, axis=0)
            self.priors[c] = X_c.shape[0] / self.n_samples

    def predict(self, X): 
        # for each sample x in the dataset X
        y_hat = [self.get_class_probability(x) for x in X]
        return np.array(y_hat)

    def get_class_probability(self, x):
        # store new posteriors for each class in a single list
        posteriors = list()

        for c in range(self.n_classes):
            # get summary stats & prior
            mean = self.mean[c]
            variance = self.variance[c]
            prior = np.log(self.priors[c])
                
            # calculate new posterior & append to list
            posterior = np.sum(np.log(self.gaussian_density(x, mean, variance)))
            posterior = prior + posterior
            posteriors.append(posterior)
        # print(self.gaussian_density(x, mean, variance))
        # return the index with the highest class probability
        return np.argmax(posteriors)
  
    def gaussian_density(self, x, mean, var):
    # implementation of gaussian density function
        const = 1 / (np.sqrt(2 * np.pi * var))
        proba = np.exp(-((x - mean) ** 2) / (2 * var))
        return const * proba
```

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_df.drop(columns = ['satisfaction']), train_df['satisfaction'], test_size=0.15, random_state=42)
```

```python
gnb = GaussianNaiveBayes(n_samples=X_train.shape[0], n_features=X_train.shape[1], n_classes=2)
```

```python
gnb.fit(X_train, y_train)
```

```python
print(classification_report(y_val, gnb.predict(np.array(X_val))))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from matplotlib import pyplot as plt
```

```python
predictions = (gnb.predict(np.array(X_val))).astype(int)
cm = confusion_matrix(y_val, predictions, labels=np.array([0, 1]))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.array(['0', '1']))
disp.plot()
plt.show()
```
